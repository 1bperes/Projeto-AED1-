```python
import random

def introducao():
    print("=" * 50)
    print("Você acorda em uma cabana velha, com a luz da lua entrando por frestas na madeira.")
    print("Sua cabeça dói e você não se lembra de como chegou aqui.")
    print("Há uma porta à sua frente e uma mesa com alguns objetos.")
    print("=" * 50)

    chave = False
    papel_lido = ""
    lapis = False
    
    while True: # O loop agora garante fazer escolhas repetidamente
        print("\nO que você faz?")
        print("\n1 - Tentar abrir a porta")
        print("2 - Investigar a mesa")
        print("3 - Fica parado tentando entender o que houve")
        
        # Pede a entrada dentro do loop para repetir a ação
        escolha = input("> ").lower()
        print("-" * 50)

        if "1" in escolha:
            if chave:
                print("Você usa a chave para destrancar a porta e sai da cabana...")
                print("A luz da lua ilumina a floresta à sua frente.")
                return chave, lapis
                break  # SAÍDA: Fim da introdução
            else:
                print("A porta está trancada, deve haver alguma chave por aqui.") 
                
        elif "2" in escolha:
            if not chave:
                
                sorteio_papel = random.randint(1, 2)
                
                print("Você encontrou a chave, uma folha de papel e uma lanterna.")
                
                chave = True
                if sorteio_papel == 1:
                    print("Na folha de papel, não há nada escrito. Você encontra um lápis ao lado.")
                    lapis = True
                else:
                    print("Na folha está escrito de forma desesperada a mensagem: NÃO CONFIE NAS VOZES!!!")
                    papel_lido = "aviso"
            else:
                print("Você já pegou tudo que tinha na mesa.")
                
        elif "3"in escolha:
            print("Você vê uma sombra se aproximando de você. Sem nada para te proteger, você é morto por ela...")
            break # SAÍDA: Final 1
            
        else:
            print(f"'{escolha}' não é uma opção válida! Tente 1, 2 ou 3.")

# Inicia a função do jogo
if __name__ == "__main__":
    introducao()
```
